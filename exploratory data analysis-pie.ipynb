```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("traffic_accidents.csv")

# Exploratory Data Analysis (EDA) - Pie Chart Context Comparison
context_col = ["lighting_condition", "crash_type", "intersection_related_i", "most_severe_injury","damage"]

for dt in context_col:
    top_context = df[dt].value_counts()
    
    # Pie Chart
    plt.figure(figsize=(8,8))
    top_context.plot.pie(autopct='%1.1f%%', startangle=90)
    plt.title(f"{dt} Comparison")
    plt.ylabel("")
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Exploratory Data Analysis (EDA) - Pie Chart Crash Time Period
df = pd.read_csv("traffic_accidents.csv")

# Categorize Time
def categorize_time(hour):
    if 0 <= hour <= 12:
        return "AM"
    elif 13 <= hour <= 23:
        return "PM"

# Group by Period and Count
df['crash_time_period'] = df['crash_hour'].apply(categorize_time)
crash_grouped = df['crash_time_period'].value_counts()

# Pie Chart
plt.figure(figsize=(8,8))
crash_grouped.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title(f"Crash Time Comparison")
plt.ylabel("")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Exploratory Data Analysis (EDA) - Pie Chart Crash Quartal Period
df = pd.read_csv("traffic_accidents.csv")

# Categorize Time
def categorize_time(month):
    if 1 <= month <= 3:
        return "Quartal 1"
    elif 4 <= month <= 6:
        return "Quartal 2"
    elif 7 <= month <= 9:
        return "Quartal 3"
    elif 10 <= month <= 12:
        return "Quartal 4"

# Group by Period and Count
df['crash_quartal_period'] = df['crash_month'].apply(categorize_time)
crash_grouped = df['crash_quartal_period'].value_counts()

# Pie Chart
plt.figure(figsize=(8,8))
crash_grouped.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title(f"Crash Quartal Comparison")
plt.ylabel("")
plt.show()
```
