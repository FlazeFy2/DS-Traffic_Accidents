```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("traffic_accidents.csv")

# Data Understanding - Variable & Data Type
print(df.dtypes)
print("\n")
print(f"{len(df.columns)} Column")
```

```python
# Data Preparation - Handle Missing Value
df.isnull().sum()
```

```python
# Data Understanding - Descriptive Statistic
df.describe()
```

```python
# Data Preparation - Outliers
numeric_columns = [
    "num_units", "injuries_total", "injuries_fatal", "injuries_incapacitating", "injuries_non_incapacitating", 
    "injuries_reported_not_evident", "injuries_no_indication", "crash_hour", "crash_day_of_week", "crash_month"
]

for dt in numeric_columns:
    plt.boxplot(df[dt])
    plt.title(f"Boxplot of {dt}")
    plt.show()
```

```python
# Data Preparation - Invalid Value
for dt in numeric_columns:
    invalid_value = df[df[dt] < 0]
    print(invalid_value.index)
```
