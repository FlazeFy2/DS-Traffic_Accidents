```python
import pandas as pd
import statistics
import numpy as np

# Declaration
# Descriptive Statistic for Analyzed Accidents by Day of the Week and Crash Date
data = pd.read_csv("traffic_accidents.csv")

# Convert Crash Day
data['crash_date'] = pd.to_datetime(data['crash_date'], format='%m/%d/%Y %I:%M:%S %p')
data['Crash Day'] = data['crash_date'].dt.day_name()

# Unique values for crash day
crash_days = data['Crash Day'].unique().tolist()

results = []
for day in crash_days:
    day_data = data[data['Crash Day'] == day]
    
    total_accident = day_data['injuries_total'].count() 
    min_injury = day_data['injuries_total'].min()  
    max_injury = day_data['injuries_total'].max()  
    avg_injury = day_data['injuries_total'].mean()  

    weather_condition = day_data['weather_condition'].mode()[0]  
    trafficway_type = day_data['trafficway_type'].mode()[0]  
    
    # Append result dictionary for each crash day
    results.append({
        'Crash Day': day,
        'Total Accident': total_accident,
        'Min Injuries Total': min_injury,
        'Max Injuries Total': max_injury,
        'Average Injuries Total': round(avg_injury, 2),
        'Most Weather Condition': weather_condition,
        'Most Traffic Way Type': trafficway_type
    })

# Sort
res_df = pd.DataFrame(results)
res_df = res_df.sort_values(by='Average Injuries Total', ascending=False)
res_df
```
